```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read in data

# define directory path
first_round_dir = '../data/simulation_data/experimental_conditions/first_round/'

# read in .txt files, values separated by whitespace
mpr_1_bound = np.loadtxt(first_round_dir + '1_mpr_bound.txt')
mpr_1_unbound = np.loadtxt(first_round_dir + '1_mpr_unbound.txt')
mpr_2_bound = np.loadtxt(first_round_dir + '2_mpr_bound.txt')
mpr_2_unbound = np.loadtxt(first_round_dir + '2_mpr_unbound.txt')
mpr_3_bound = np.loadtxt(first_round_dir + '3_mpr_bound.txt')
mpr_3_unbound = np.loadtxt(first_round_dir + '3_mpr_unbound.txt')

mprf_1_bound = np.loadtxt(first_round_dir + '1_mprf_bound.txt')
mprf_1_unbound = np.loadtxt(first_round_dir + '1_mprf_unbound.txt')
mprf_2_bound = np.loadtxt(first_round_dir + '2_mprf_bound.txt')
mprf_2_unbound = np.loadtxt(first_round_dir + '2_mprf_unbound.txt')
mprf_3_bound = np.loadtxt(first_round_dir + '3_mprf_bound.txt')
mprf_3_unbound = np.loadtxt(first_round_dir + '3_mprf_unbound.txt')

rl_1 = np.loadtxt(first_round_dir + '1_read_lengths.txt')
rl_2 = np.loadtxt(first_round_dir + '2_read_lengths.txt')
rl_3 = np.loadtxt(first_round_dir + '3_read_lengths.txt')
```

```python
# print total number of reads per pool
print('Total number of reads per pool:')
print('Pool 1: ' + str(len(mpr_1_bound) + len(mpr_1_unbound)))
print('Pool 2: ' + str(len(mpr_2_bound) + len(mpr_2_unbound)))
print('Pool 3: ' + str(len(mpr_3_bound) + len(mpr_3_unbound)))

# print total number of reads per pool after filtering
print('Total number of reads per pool after filtering:')
print('Pool 1: ' + str(len(mprf_1_bound) + len(mprf_1_unbound)))
print('Pool 2: ' + str(len(mprf_2_bound) + len(mprf_2_unbound)))
print('Pool 3: ' + str(len(mprf_3_bound) + len(mprf_3_unbound)))
```

```python
# plot histograms of read lengths
# color the bins < 30 red
fig, ax = plt.subplots(2, 2, figsize=(8,8))
ax[0, 0].hist(rl_1, bins=100, color='blue')
ax[0, 0].set_title('Pool 1')
ax[0, 0].set_xlabel('Read Length')
ax[0, 0].set_ylabel('Count')
ax[0, 0].axvline(x=30, color='red')
ax[0, 1].hist(rl_2, bins=100, color='blue')
ax[0, 1].set_title('Pool 2')
ax[0, 1].set_xlabel('Read Length')
ax[0, 1].set_ylabel('Count')
ax[0, 1].axvline(x=30, color='red')
ax[1, 0].hist(rl_3, bins=100, color='blue')
ax[1, 0].set_title('Pool 3')
ax[1, 0].set_xlabel('Read Length')
ax[1, 0].set_ylabel('Count')
ax[1, 0].axvline(x=30, color='red')

plt.tight_layout()
```

```python
# plot histograms of mutations per read bound vs unbound
fig, ax = plt.subplots(2, 2, figsize=(8,8))
ax[0, 0].hist(mpr_1_bound, bins=10, color='blue', alpha=0.5, label='Bound')
ax[0, 0].hist(mpr_1_unbound, bins=10, color='red', alpha=0.5, label='Unbound')
ax[0, 0].set_title('Pool 1')
ax[0, 0].set_xlabel('Mutations per Read')
ax[0, 0].set_ylabel('Count')
ax[0, 0].legend()
ax[0, 1].hist(mpr_2_bound, bins=10, color='blue', alpha=0.5, label='Bound')
ax[0, 1].hist(mpr_2_unbound, bins=10, color='red', alpha=0.5, label='Unbound')
ax[0, 1].set_title('Pool 2')
ax[0, 1].set_xlabel('Mutations per Read')
ax[0, 1].set_ylabel('Count')
ax[0, 1].legend()
ax[1, 0].hist(mpr_3_bound, bins=10, color='blue', alpha=0.5, label='Bound')
ax[1, 0].hist(mpr_3_unbound, bins=10, color='red', alpha=0.5, label='Unbound')
ax[1, 0].set_title('Pool 3')
ax[1, 0].set_xlabel('Mutations per Read')
ax[1, 0].set_ylabel('Count')
ax[1, 0].legend()

plt.tight_layout()
```

```python
# plot histograms of mutations per read bound vs unbound after filtering
fig, ax = plt.subplots(2, 2, figsize=(8,8))
ax[0, 0].hist(mprf_1_bound, bins=10, color='blue', alpha=0.5, label='Bound')
ax[0, 0].hist(mprf_1_unbound, bins=10, color='red', alpha=0.5, label='Unbound')
ax[0, 0].set_title('Pool 1')
ax[0, 0].set_xlabel('Mutations per Read')
ax[0, 0].set_ylabel('Count')
ax[0, 0].legend()
ax[0, 1].hist(mprf_2_bound, bins=10, color='blue', alpha=0.5, label='Bound')
ax[0, 1].hist(mprf_2_unbound, bins=10, color='red', alpha=0.5, label='Unbound')
ax[0, 1].set_title('Pool 2')
ax[0, 1].set_xlabel('Mutations per Read')
ax[0, 1].set_ylabel('Count')
ax[0, 1].legend()
ax[1, 0].hist(mprf_3_bound, bins=10, color='blue', alpha=0.5, label='Bound')
ax[1, 0].hist(mprf_3_unbound, bins=10, color='red', alpha=0.5, label='Unbound')
ax[1, 0].set_title('Pool 3')
ax[1, 0].set_xlabel('Mutations per Read')
ax[1, 0].set_ylabel('Count')
ax[1, 0].legend()

plt.tight_layout()
```
