```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read in data

# define directory path
first_round_dir = '../data/simulation_data/first_round/'

# read in .txt files, values separated by whitespace
mpr_1 = np.loadtxt(first_round_dir + '1_mutations_per_read.txt')
mpr_2 = np.loadtxt(first_round_dir + '2_mutations_per_read.txt')
mpr_3 = np.loadtxt(first_round_dir + '3_mutations_per_read.txt')
mpr_4 = np.loadtxt(first_round_dir + '4_mutations_per_read.txt')

mprf_1 = np.loadtxt(first_round_dir + '1_mutations_per_read_filtered.txt')
mprf_2 = np.loadtxt(first_round_dir + '2_mutations_per_read_filtered.txt')
mprf_3 = np.loadtxt(first_round_dir + '3_mutations_per_read_filtered.txt')
mprf_4 = np.loadtxt(first_round_dir + '4_mutations_per_read_filtered.txt')

rl_1 = np.loadtxt(first_round_dir + '1_read_lengths.txt')
rl_2 = np.loadtxt(first_round_dir + '2_read_lengths.txt')
rl_3 = np.loadtxt(first_round_dir + '3_read_lengths.txt')
rl_4 = np.loadtxt(first_round_dir + '4_read_lengths.txt')
```

```python
# print total number of reads per pool
print('Total number of reads per pool:')
print('Pool 1: ' + str(len(mpr_1)))
print('Pool 2: ' + str(len(mpr_2)))
print('Pool 3: ' + str(len(mpr_3)))
print('Pool 4: ' + str(len(mpr_4)))

# print total number of reads per pool after filtering
print('\nTotal number of reads per pool after filtering:')
print('Pool 1: ' + str(len(mprf_1)))
print('Pool 2: ' + str(len(mprf_2)))
print('Pool 3: ' + str(len(mprf_3)))
print('Pool 4: ' + str(len(mprf_4)))
```

```python
# plot histograms of mutations per read
fig, ax = plt.subplots(2, 2, figsize=(10, 10))
ax[0, 0].hist(mpr_1, bins=range(0, 10), align='left', rwidth=0.8)
ax[0, 0].set_title('B = 1')
ax[0, 0].set_xlabel('Mutations Per Read')
ax[0, 0].set_ylabel('Frequency')
ax[0, 1].hist(mpr_2, bins=range(0, 10), align='left', rwidth=0.8)
ax[0, 1].set_title('B = 6')
ax[0, 1].set_xlabel('Mutations Per Read')
ax[0, 1].set_ylabel('Frequency')
ax[1, 0].hist(mpr_3, bins=range(0, 10), align='left', rwidth=0.8)
ax[1, 0].set_title('B = 15')
ax[1, 0].set_xlabel('Mutations Per Read')
ax[1, 0].set_ylabel('Frequency')
ax[1, 1].hist(mpr_4, bins=range(0, 10), align='left', rwidth=0.8)
ax[1, 1].set_title('B = 30')
ax[1, 1].set_xlabel('Mutations Per Read')
ax[1, 1].set_ylabel('Frequency')
plt.tight_layout()
```

```python
# plot histograms of mutations per read (filtered)
fig, ax = plt.subplots(2, 2, figsize=(10, 10))
ax[0, 0].hist(mprf_1, bins=range(0, 10), align='left', rwidth=0.8)
ax[0, 0].set_title('B = 1 (Filtered)')
ax[0, 0].set_xlabel('Mutations Per Read')
ax[0, 0].set_ylabel('Frequency')
ax[0, 1].hist(mprf_2, bins=range(0, 10), align='left', rwidth=0.8)
ax[0, 1].set_title('B = 6 (Filtered)')
ax[0, 1].set_xlabel('Mutations Per Read')
ax[0, 1].set_ylabel('Frequency')
ax[1, 0].hist(mprf_3, bins=range(0, 10), align='left', rwidth=0.8)
ax[1, 0].set_title('B = 15 (Filtered)')
ax[1, 0].set_xlabel('Mutations Per Read')
ax[1, 0].set_ylabel('Frequency')
ax[1, 1].hist(mprf_4, bins=range(0, 10), align='left', rwidth=0.8)
ax[1, 1].set_title('B = 30 (Filtered)')
ax[1, 1].set_xlabel('Mutations Per Read')
ax[1, 1].set_ylabel('Frequency')
plt.tight_layout()
```

```python
# plot histograms of read lengths
fig, ax = plt.subplots(2, 2, figsize=(10, 10))
ax[0, 0].hist(rl_1, bins=range(0, 100), align='left', rwidth=0.8)
ax[0, 0].set_title('B = 1')
ax[0, 0].set_xlabel('Read Length')
ax[0, 0].set_ylabel('Frequency')
ax[0, 1].hist(rl_2, bins=range(0, 100), align='left', rwidth=0.8)
ax[0, 1].set_title('B = 6')
ax[0, 1].set_xlabel('Read Length')
ax[0, 1].set_ylabel('Frequency')
ax[1, 0].hist(rl_3, bins=range(0, 100), align='left', rwidth=0.8)
ax[1, 0].set_title('B = 15')
ax[1, 0].set_xlabel('Read Length')
ax[1, 0].set_ylabel('Frequency')
ax[1, 1].hist(rl_4, bins=range(0, 100), align='left', rwidth=0.8)
ax[1, 1].set_title('B = 30')
ax[1, 1].set_xlabel('Read Length')
ax[1, 1].set_ylabel('Frequency')
plt.tight_layout()
```
