# figures for master thesis

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
```

```python
# example figure for linear regression
# set figure size
plt.figure(figsize=(8, 8))
n = 30
#uniform random draws for x between 0 and 10
x = np.random.rand(n)*10
y = 2*x + 1 + np.random.randn(n)
plt.scatter(x, y)
plt.xlabel('x')
plt.ylabel('y')
# add regression line
plt.plot([0, 10], [1, 21], 'black')
# add equation
plt.text(1, 20, r'$y = wx + b$', fontsize=15)
# add w and b
plt.text(1, 18, r'$w = 2$', fontsize=15)
plt.text(1, 16, r'$b = 1$', fontsize=15)
# add legend bottom right
plt.legend(['data','regression line'], fontsize=15, loc='lower right')
plt.show()
```

```python
# example figure for logistic regression
# set figure size
plt.figure(figsize=(8, 8))
n = 30
# generate binary data
y0 = np.random.randn(n) + 2
y1 = np.random.randn(n) + 6
#y0 is class 0, y1 is class 1
x = np.concatenate((y0, y1))
y = np.concatenate((np.zeros(n), np.ones(n)))
plt.scatter(x, y)
plt.xlabel('x')
plt.ylabel('y')
# fit logistic regression
lr = LogisticRegression()
lr.fit(x.reshape(-1, 1), y)
# add logistic regression curve to plot
x_test = np.linspace(0, 10, 100)
y_test = lr.predict_proba(x_test.reshape(-1, 1))[:, 1]
plt.plot(x_test, y_test, 'black')
# add equation
plt.text(1, 0.8, r'$y = \frac{1}{1 + e^{-(wx + b)}}$', fontsize=15)
# add w and b
plt.text(1, 0.7, r'$w = 2.5$', fontsize=15)
plt.text(1, 0.6, r'$b = -10$', fontsize=15)
# add legend bottom right
plt.legend(['data','logistic regression'], fontsize=15, loc='lower right')
plt.show()
```

```python
#example figure derivative as slope
# set figure size
plt.figure(figsize=(8, 8))
# quadratic function with tangent line
x = np.linspace(-5, 5, 100)
y = x**2
x0 = 2
y0 = x0**2
# tangent line
x1 = np.linspace(0, 4, 100)
y1 = 4*x1 - 4
plt.plot(x, y)
plt.plot(x1, y1, 'black')
plt.scatter(x0, y0, s=100)
plt.xlabel('w')
plt.ylabel('loss')
#add equation for function
plt.text(1, 22, r'$L(w) = w^2$', fontsize=15)
# add equation for derivative
plt.text(1, 20, r'$\frac{dL}{dw} = 2w$', fontsize=15)
# add w and b
plt.text(1, 18, r'$w = 2$', fontsize=15)
plt.text(1, 16, r'$\frac{dL}{dw} = 4$', fontsize=15)
# add legend bottom right
plt.legend(['function','tangent'], fontsize=15, loc='lower right')
plt.show()
```

```python
# example loss function with w and b as parameters
# 3d plot
# set figure size
plt.figure(figsize=(8, 8))
def Loss(x_plot, y_plot):
    return (x_plot - 2)**2 + (y_plot - 3)**2
sns.set_style('whitegrid')
num = 40
w = np.linspace(0, 10, num)
b = np.linspace(0, 10, num)
plot1, plot2 = np.meshgrid(w, b)
plot3 = Loss(plot1, plot2)
axes = plt.axes(projection='3d')
axes.plot_surface(plot1, plot2, plot3, cmap='viridis', alpha=0.5)
# add minimum as point in 3d plot
axes.scatter(2, 3, Loss(2, 3), color='red', s=100)
plt.show()
```

```python
# example figure bias variance tradeoff
# set figure size
plt.figure(figsize=(8, 8))
# set seed
np.random.seed(3)
# quadratic function
x = np.linspace(0, 5, 100)
y = x**2
plt.plot(x, y)
# add random collection of points
x1 = np.random.rand(10)*5
y1 = x1**2 + np.random.randn(10)*2
plt.scatter(x1, y1)
plt.xlabel('x')
plt.ylabel('y')
# add equation for function
plt.text(1, 22, r'$y = x^2$', fontsize=15)
# add legend bottom right
plt.legend(['function','data'], fontsize=15, loc='lower right')
plt.show()
```

```python
# fiogure of ReLU
# set figure size
plt.figure(figsize=(8, 8))
# ReLU function
x = np.linspace(-5, 5, 100)
y = np.maximum(0, x)
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
# add equation for function
plt.text(1, 4.5, r'$y = max(0, x)$', fontsize=15)
plt.show()
```

```python
# figure for exponentiallly weighted moving average
# set figure size
plt.figure(figsize=(12, 8))
# get time series data
x = np.linspace(0, 10, 100)
y = np.sin(x)
# add noise
y = y + np.random.randn(100)*0.5
# plot data
plt.plot(x, y)
# add exponentiallly weighted moving average with beta = 0.9
y_ewma = np.zeros(100)
y_ewma[0] = y[0]
for i in range(1, 100):
    y_ewma[i] = 0.9*y_ewma[i-1] + 0.1*y[i]
plt.plot(x, y_ewma)
# add exponentiallly weighted moving average with beta = 0.99
y_ewma = np.zeros(100)
y_ewma[0] = y[0]
for i in range(1, 100):
    y_ewma[i] = 0.99*y_ewma[i-1] + 0.01*y[i]
plt.plot(x, y_ewma)
# add legend bottom right
plt.legend(['data','ema with beta = 0.9', 'ema with beta = 0.99'], fontsize=15, loc='lower right')
plt.show()

```
